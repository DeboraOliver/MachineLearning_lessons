```python
from numpy import loadtxt
import numpy as np
from keras.models import Sequential
from keras.layers import Dense #keras é uma api dentro do tensor flow
```

```python
#carregar dataset sobre indios com diabetes

dataset = loadtxt('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv', delimiter=',')

X = dataset[:,0:8]
y = dataset[:,8]

#instancio o modelo como sendo do tipo sequencial
model = Sequential()

#1º camada do tipo dense, relu que é uma função linear. A saída será positiva se a entrada for positiva (a mesma), mas será 0 se a entrada for negativa
model.add(Dense(12, input_dim = 8,  activation = 'relu')) 
#12 são 12 neurônios artificiasi e 8 as 8 caracteristicas

#2º camada para definir o modelo
model.add(Dense(8, activation = 'relu'))

#3ª camada
model.add(Dense(1, activation = 'sigmoid'))
#o que é e pq é sigmoide
#1 unico nerunio consegue definir funções nos quais a saída são 01 ou 1
```

```python
model.compile(loss = 'binary_crossentropy', optimizer= 'adam', metrics = ['accuracy']) 
#a função de perda loss, o optimizador, adam é um gradiente descendente
```

```python
#treinar o modelo

model.fit(X,y, epochs= 150, batch_size= 10) 
#epocas são lotes, compostas por uma ou mais amostras. 10 lotes para treinar 150 epocas.
```

```python
_, accuracy = model.evaluate(X,y) #o evaluate retorna a perda e a acuracia, deixei o _ porque embora ele forneça a loss (perda) eu não quero
print("Acurácia : ", (accuracy*100))

#veja que a soma da loss com a acurácia dá valor maior que 100
# pq a perda é 46% é alto, precisamos melhorar a acuracia
```
