```python

""" Parte 1: Regressão

Você deve criar um modelo de regressão linear para predição da chance de diabetes em pessoas a partir de algumas características: idade, sexo, massa corpórea, pressão arterial, dentre outras. 
Essas características descritas anteriormente serão os dados de entrada do algoritmo, elas já foram coletadas e estão disponíveis no banco de dados “diabetes” da biblioteca Scikit-learn

Você deve, então, dividir os dados em dados para treino e dados para teste do modelo de predição, usando a função “train_test_split()” da biblioteca Scikit-learn​. 
Você deve definir que somente 10% dos dados serão usados para teste e que o valor de aleatoriedade é 23 (random_state = 23) para que possa comparar o seu resultado com os da atividade.

Na sequência, crie o modelo linear usando o pacote “linear_model” da biblioteca Scikit-learn​ e então realize a predição para os dados que foram separados para teste pelo “train_test_split()”.

O resultado gerado pelo regressor (predições) deve ser igual a imagem ​"""
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df["resultado"] = diabetes.target 

df.head(5)
```

```python
#Cria o dataframe com as características que você selecionou.
x = df.iloc[:,0:10].values

#Define a variável de saída.

Y = df["resultado"]

#Realiza a divisão dos dados em treino e teste.

x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size = 0.1, random_state=23)

#Para predizer a sobrevida do paciente.

model = LinearRegression()

model.fit(x_train, Y_train)

sobrevida_paciente = model.predict(x_test)

print (sobrevida_paciente)
```

```python
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_digits

digitos = load_digits()

df = pd.DataFrame(digitos.data, columns=digitos.feature_names)
df["resultado"] = digitos.target 

df.head(5)
```

```python
x_train, x_test, y_train, y_test = train_test_split(digitos.data, digitos.target, test_size=0.30, random_state=50)

regressaologistica = LogisticRegression()

regressaologistica.fit(x_train, y_train)

score = regressaologistica.score(x_test, y_test)

print("Score: ", score)
```
